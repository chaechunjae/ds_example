```python
import seaborn as sns
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset("penguins").dropna()
penguins
```

```python
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm", errorbar=None)
```

```python
sns.lineplot(data=penguins, x="body_mass_g", y="bill_length_mm", errorbar=None, hue="species")
```

```python
sns.pointplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette="Set2")
```

```python
sns.histplot(data=penguins, x="bill_depth_mm", bins=30, hue="species", multiple="stack")
```

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex")
```

```python
sns.countplot(data=penguins, y="sex")
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g")
```

```python
sns.swarmplot(data=penguins, x="sex", y="flipper_length_mm", palette="summer")
```

```python
sns.scatterplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="species")
```
