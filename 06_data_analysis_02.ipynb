```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apple = pd.read_excel("./naver_shopping_애플.xlsx")

apple["published_date"]  = apple["published_date"] + "01"
apple["published_date"] = pd.to_datetime(apple["published_date"])

apple
```

```python
from collections import Counter

product_list = []
apple["Product Name"].str.split().tolist()
for row in apple["Product Name"].str.split().tolist():
    product_list += row

product_list
Counter(product_list).most_common(30)
```
